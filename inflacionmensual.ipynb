```python
import pandas as pd
import matplotlib.pyplot as plt

# Dataset: Inflaci√≥n mensual Argentina (%)
data = {
    "Mes" : ["2024-01","2024-02","2024-03","2024-04","2024-05","2024-06",
             "2024-07","2024-08","2024-09","2024-10","2024-11","2024-12"
            ],
        "Inflaci√≥n" : [20.6, 13.2, 11.0, 8.8, 6.5, 4.6, 3.5, 4.2, 5.4, 6.3, 7.0, 8.0]
}
df = pd.DataFrame(data)

#Ver los primeros datos
print(df.head())

#Gr√°fico
plt.figure(figsize=(10,5))
plt.plot(df["Mes"], df["Inflaci√≥n"], marker="o")
plt.title("Inflaci√≥n Mensual Argentina 2024")
plt.xlabel("Mes")
plt.ylabel("Inflaci√≥n (%)")
plt.xticks(rotation=45)
plt.grid(True)
plt.show()
```

```python
#Promedio anual de inflaci√≥n
promedio = df["Inflaci√≥n"].mean()
print("Promedio anual de inflaci√≥n:", round(promedio,2), "%")

#Mes con inflaci√≥n m√°s alta y m√°s baja
max_mes = df.loc[df["Inflaci√≥n"].idxmax()]
min_mes = df.loc[df["Inflaci√≥n"].idxmin()]

print("Mayor inflaci√≥n:", max_mes["Inflaci√≥n"], "% en", max_mes["Mes"])
print("Menor inflaci√≥n:", min_mes["Inflaci√≥n"], "% en", min_mes["Mes"])

#Inflaci√≥n acumulada
df["Acumulada"] = df["Inflaci√≥n"].cumsum()
print("\nDatos con inflaci√≥n acumulada:")
print(df)

plt.figure(figsize=(10,5))
plt.bar(df["Mes"], df["Inflaci√≥n"])
plt.title("Inflaci√≥n Mensual - Barras")
plt.xlabel("Mes")
plt.ylabel("Inflaci√≥n (%)")
plt.xticks(rotation=45)
plt.grid(axis="y")
plt.show
```

```python
# Grafico comparativo: l√≠nea mensual + l√≠nea acumulada
plt.figure(figsize=(12,6))

plt.plot(df["Mes"], df["Inflaci√≥n"], marker="o", label="Mensual")
plt.plot(df["Mes"], df["Acumulada"], marker="s", label="Acumulada")

plt.title("Inflaci√≥n Argentina 2024 - Mensual vs. Acumulada")
plt.xlabel("Mes")
plt.ylabel("Inflaci√≥n (%)")
plt.xticks(rotation=45)
plt.grid(True)
plt.legend()

plt.show()
```
